```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import en_core_web_sm

# Cargar modelo en español de spaCy
nlp = en_core_web_sm.load()
```

```python
df = pd.read_csv("tweets_filtered21.csv")
```

```python
# Función para verificar si el tweet contiene una palabra clave o sus variantes
def contains_keyword(text, keywords, nlp):
    doc = nlp(text)
    lemmas = {token.lemma_ for token in doc}
    return any(keyword in lemmas for keyword in keywords)
```

```python
keywords0 = [] #Reportes informes
keywords1 = [] #Especulaciones, prevencion
keywords2 = [] #Existencia de daño
keywords3 = [] #Daño ocurrido
keywords4 = [] #Daño estructural o muerte
import re

# Definir una función para clasificar los tweets usando expresiones regulares (regex)
def clasificar_tweet_regex(texto):
    texto = texto.lower()
    # Nivel 3: Daños graves o incidentes mayores
    if re.search(r'\b(daños graves|deslaves|colapso|accidente grave|muertes|daños estructurales)\b', texto):
        return 3
    # Nivel 2: Accidentes moderados o cierres importantes
    elif re.search(r'\b(accidente|colisión|choque|cierres parciales|vehículos involucrados)\b', texto):
        return 2
    # Nivel 1: Incidentes menores o advertencias
    elif re.search(r'\b(incendio|fuerte lluvia|alerta|precaución)\b', texto):
        return 1
    # Nivel 0: Sin daños significativos
    else:
        return 0

# Aplicar la clasificación a los tweets sin nivel asignado
df['Nivel'] = df['Texto'].apply(clasificar_tweet_regex)
```
